```python
INSIGHTS:-
Here we are getting the summary of the statistics
- Average total bill is around 19 → most customers spend a moderate amount.
- Average tip is around 3 → tipping is fairly consistent.
- Bills vary a lot, tips vary less → people tip in a predictable range.
- Average group size is 2–3 → mostly small groups or couples.
- No missing values in numeric columns → data is clean
```

```python
INSIGHTS:-
1. Most customers spend a moderate amount
- The average total bill is around 19.
- Bills range from very small (around 3) to quite large (around 50).
2. Tipping behavior is fairly consistent
- Average tip is around 3.
- Tips don’t vary as much as total bills, meaning people tip in a predictable range.
3. Small groups dominate
- Average group size is 2–3 people.
- Maximum size is 6, so large groups are rare.
4. Higher bills usually lead to higher tips
- Since both max bill and max tip are high, customers who spend more tend to tip more.
5. Dataset is clean
- No missing values in numeric columns.
- Ready for analysis without cleaning.
```
